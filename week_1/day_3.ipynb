# Data cleaning continued

```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv('./data/cleaned_data.csv')

# Remove leading/trailing whitespace and replace non-breaking spaces
df.columns = df.columns.str.strip().str.replace('\xa0', ' ').str.replace('  ', ' ')
print(df.columns.tolist())

# Standardizing numerical columns with currency symbols
def clean_currency_column(series):
    return pd.to_numeric(series.replace(r'[\$,]', '', regex=True), errors='coerce')

# Apply the function to the appropriate columns
for col in ['Actual gross', 'Adjusted gross (in 2022 dollars)', 'Average gross']:
    df[col] = clean_currency_column(df[col])

#Drop both All Time Peak and Peak columns
df.drop(columns=['Peak', 'All Time Peak'], inplace=True)


df
```

```python
df.to_csv('./data/cleaned_data.csv', index=False)
```
