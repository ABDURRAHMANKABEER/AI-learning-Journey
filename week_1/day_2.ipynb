# ðŸ”§ Step 3: Techniques to Handle Missing Data
1. Drop Rows or Columns
Use when the data is mostly missing or not important.

```python
df_cleaned = df.dropna()
df_cleaned
```

# 2. Fill with a Default Value
Use when it makes sense to replace missing with 0 or "Unknown"

```python
df['All Time Peak'] = df['All Time Peak'].fillna('Unknown')
df['Peak'] = df['Peak'].fillna('Unknown')
df
```

```python
cleaned_data = df.to_csv('./data/cleaned_data.csv', index=False)
```
